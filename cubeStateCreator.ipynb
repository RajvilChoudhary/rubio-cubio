```python
import random
from kociemba.pykociemba import cubiecube

for _ in range(10):
    cube = cubiecube.CubieCube()

    for _ in range(25):
        face = random.randint(0, 5)      # U,R,F,D,L,B
        turns = random.randint(1, 3)     # 90掳,180掳,270掳
        for _ in range(turns):
            cube.multiply(cubiecube.moveCube[face])

    face_cube = cube.toFaceCube()
    print(face_cube.to_String())
```

```python
from kociemba.pykociemba import cubiecube, facecube

# -------------------------------
# INPUT
# -------------------------------
initial_state = "UUUUUUUUURRRRRRRRRFFFFFFFFFDDDDDDDDDLLLLLLLLLBBBBBBBBB"

moves = "R L U2 R L F2 R2 U2 R2 F2 R2 U2 F2 L2"

# -------------------------------
# Convert facelets -> CubieCube
# -------------------------------
fc = facecube.FaceCube(initial_state)
cube = fc.toCubieCube()

# Mapping
move_map = {
    'U': 0,
    'R': 1,
    'F': 2,
    'D': 3,
    'L': 4,
    'B': 5
}

# Apply each move
for mv in moves.split():
    face = move_map[mv[0]]

    if len(mv) == 1:
        turns = 1          # 90掳
    elif mv[1] == '2':
        turns = 2          # 180掳
    elif mv[1] == "'":
        turns = 3          # 270掳 = inverse
    else:
        raise ValueError(f"Invalid move: {mv}")

    for _ in range(turns):
        cube.multiply(cubiecube.moveCube[face])

# Convert back to facelets
final_state = cube.toFaceCube().to_String()

print("Final state:")
print(final_state)
```
